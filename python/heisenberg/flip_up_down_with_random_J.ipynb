```python
energy_list = []
config_list = []
J_list = []
```

```python
energy_list = []
config_list = []
J_list = []
for i in range(1000):
    model = Up_down_flip_model(N=4,step=500,seed=i,label=''+str(i))
    model.simulation(temperature=1.5)
    energy_list.append(model.lowest_energy)
    config_list.append(model.lowest_config)
    J_list.append(model.alpha)
    print('step: ',str(i))


```

```python
energy = [ i[np.newaxis,...] for i in energy_list]
energy = np.concatenate(energy)
with open('energy.npy', 'wb') as f:
    np.save(f,energy)
```

```python
config = [i.flatten()[np.newaxis,...] for i in config_list]
# expexted shape = [num,2,4,4]
config = np.concatenate(config)
with open('config.npy', 'wb') as f:
    np.save(f,config)
```

```python
J = [i.flatten()[np.newaxis,...] for i in J_list]
# expexted shape = [num,2,4,4]
J = np.concatenate(J,axis = 0)
with open('J.npy', 'wb') as f:
    np.save(f,J)
```
